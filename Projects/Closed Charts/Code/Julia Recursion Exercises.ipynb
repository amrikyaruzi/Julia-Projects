## Fibonacci

```julia
function fibonacci(n::Int)::BigInt
    memo = Dict{Int, BigInt}()
    return _fibonacci(n, memo)
end

function _fibonacci(n::Int, memo::Dict)::BigInt
    if n == 0
        return 0
    elseif n == 1 || n == 2
        return 1
    elseif haskey(memo, n)
        return memo[n]
    else
        fib_value = _fibonacci(n-1, memo) + _fibonacci(n-2, memo)
        memo[n] = fib_value
        return fib_value
    end
end
```

```julia
map(fibonacci, range(0, 500))
```

```julia
fibonacci(500)
```
